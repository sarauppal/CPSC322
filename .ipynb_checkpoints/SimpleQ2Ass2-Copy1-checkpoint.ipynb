```python
import numpy as np
```

## Initialize Domains

```python

A=[1,2,3,4]
B=[1,2,3,4]
C=[1,2,3,4]
D=[1,2,3,4]
E=[1,2,3,4]
F=[1,2,3,4]
G=[1,2,3,4]
H=[1,2,3,4]

    
def consistent(V,Constraints):
    for constraint in Constraints:
        if not eval(constraint):
            return False
            break
        else:
            return True
        
            



    
```

## Define Constraints as dict

```python
Constraints={'a':[],
             'b':[],
             'c':[],
             'd':['d!=c'],
             'e':['e!=c','e<(d-1)'],
             'f':['abs(f-b)==1','c!=f','d!=f','abs(e-f)%2!=0'],
             'g':['a>=g','abs(g-c)==1','d>=g','g!=f'],
             'h':['a<h','g<h','abs(h-c)%2==0','h!=d','e!=(h-2)','h!=f']}
```

```python
assignment={}
models=np.array([])
space='\t'
for a in A:
    assignment['a']=a
    print('\n a={0}'.format(a),end="")
    time=1
    for b in B:
        assignment['b']=b
        print(' \n {0} b={1}'.format(space*time,b),end="")
        time+=1
        for c in C:
            print(' \n {0} c={1}'.format(space*time,c) ,end="")
            time+=1
            for d in D:
                print(' \n {0} d={1}'.format(space*time,d), end="")
                time+=1
                for constraint in Constraints['d']:
                    if not eval(constraint):
                        print(' \n {0}  -> failure '.format(space*time))
                        time-=1
                        flag=1
                        break
                if flag:
                    flag=0
                    continue
                
                for e in E:
                    print('\n {0} e={1}'.format(space*time,e),end="")
                    time+=1
                    for constraint in Constraints['e']:
                        if not eval(constraint):
                            print('\n {0} -> failure'.format(space*time))
                            time-=1
                            flag=1
                            break
                    if flag:
                        flag=0
                        continue
                        
                            
                    for f in F:
                        print('\n {0} f={1}'.format(space*time,f),end="")
                        time+=1
                        for constraint in Constraints['f']:
                            if not eval(constraint):
                                print('\t  -> failure'.format(f))
                                time-=1
                                flag=1
                                break
                        if flag:
                            flag=0
                            continue  
                            
                        for g in G:
                                print('\t g={0}'.format(g),end="")
                                for constraint in Constraints['g']:
                                    if not eval(constraint):
                                        print('\t -> failure'.format(g))
                                        flag=1
                                        break  
                                        
                                if flag:
                                    flag=0
                                    continue
                                    
                                    
                                for h in H:
                                    print('\t h={0}'.format(h),end="")
                                    for constraint in Constraints['h']:
                                        if not eval(constraint):
                                            print('\t  -> failure'.format(h))
                                            flag=1
                                            break
                                            
                                    if flag:
                                        flag=0
                                        continue
                                                                            
                                    print("Yay! we have a model")
                                    model=np.array([a,b,c,d,e,f,g,h])
                                    if not models.size:
                                        models=model
                                    else:
                                        models=np.vstack((models,model))
                                                                        
                                                                
                                                            
                                                
                                            
                                            
                                            
                                            
                                            
```
